```python
import re
import string

def remove_punctuation(text):
    translator_table=str.maketrans(' ',' ',string.punctuation)
    return text.translate(translator_table)
```

```python
def remove_url1(text):
    url = re.compile(r'https7://\S+|www\.\S+')
    return url.sub(r' ', text)
```

```python
df['text']=df['text'].map(remove_url1)
df['text']=df['text'].map(remove_punctuation)
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # Download stopwords data

stop = set(stopwords.words('english'))  # Set stop to English stopwords
```

```python
stop=set(stopwords.words('english'))
```

```python
def remove_sw(text):
  filtered_word=[word.lower() for word in text.split() if word.lower not in stop]
  return '  '.join(filtered_word)
```

```python
df['text']=df['text'].map(remove_sw)
```

```python
from collections import Counter

def countword(text):
    count = Counter()
    if isinstance(text, pd.Series):
        text_values = text.values
    elif isinstance(text, np.ndarray):
        if text.ndim == 1:
            text_values = text
        else:
            raise ValueError("Invalid data type. If 'text' is a NumPy array, it should be 1-dimensional.")
    else:
        raise ValueError("Invalid data type. 'text' should be a pandas Series or NumPy array.")
    for t in text.values:
        for word in t.split():
            count[word] += 1
    return count
```

```python

def countword(column):
    count = Counter()
    
    if isinstance(column, pd.Series):
        column_values = column.values
    elif isinstance(column, np.ndarray):
        if column.ndim == 1:
            column_values = column
        else:
            raise ValueError("Invalid data type. If 'column' is a NumPy array, it should be 1-dimensional.")
    else:
        raise ValueError("Invalid data type. 'column' should be a pandas Series or NumPy array.")
    
    for value in column_values:
        if pd.notnull(value):  # Exclude NaN values if any
            for word in value.split():
                count[word] += 1
    
    return count
```

```python
counter=countword(df.text)
```

```python
no_unique=len(counter)
```

```python
train=df.iloc[:6000,:]
```

```python
test=df.iloc[6000:,:]
```

```python
X_train=train.text.to_numpy()
X_test=test.text.to_numpy()
y_train=train.target.to_numpy()
y_test=train.target.to_numpy()
```

```python
X_train
```

```python
y_train
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
token=Tokenizer(num_words=no_unique)
token.fit_on_texts(X_train)
```

```python
word_index=token.word_index
```

```python
word_index
```

```python
X_train_seq=token.texts_to_sequences(X_train)
X_test_seq=token.texts_to_sequences(X_test)
```

```python
X_train
```

```python
X_train_seq
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
Max_len=17

X_train_pad=pad_sequences(X_train_seq,maxlen=Max_len,padding='post',truncating='post')
X_test_pad=pad_sequences(X_test_seq,maxlen=Max_len,padding='post',truncating='post')
```

```python
X_train_pad.shape
```

```python
X_test_pad.shape
```

```python
from tensorflow.keras import layers
```

```python
model=ke.models.Sequential()
model.add(layers.Embedding(no_unique,32,input_length=Max_len))
model.add(layers.LSTM(64,dropout=0.01))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
loss=ke.losses.BinaryCrossentropy(from_logits=False)
optime=ke.optimizers.Adam(lr=0.001)
model.compile(loss=loss,optimizer=optime,metrics=['accuracy'])
```

```python
model.fit(X_train_pad,y_train,epochs=20,verbose=2)
```

```python
prediction=model.predict(X_train_pad,)
prediction=[1 if p>0.5 else 0 for p in prediction]
```
